```python
import torch.nn.functional as F
from PIL import Image
import numpy as np

# pred = model(imgc.unsqueeze(0)).squeeze()
pred = model(imgc.unsqueeze(0)).squeeze()
pred = F.sigmoid(pred)
# pred[pred < 0.5] = 0
# pred[pred >= 0.5] = 1
# pred[pred > 0] = 1
pred = T.functional.to_pil_image(pred).resize((512, 512), resample=Image.NEAREST)
# pred = Image.fromarray(np.array(pred))
# pred.size()
```

```python
pred = np.array(pred)
pred[pred > 0] = 255
pred = Image.fromarray(pred)
pred = pred.convert('L').convert('RGBA')
datas = pred.getdata()

newData = []
for item in datas:
    if item[0] == 0 and item[1] == 0 and item[2] == 0:
        newData.append((255, 255, 255, 0))
    else:
        newData.append((0, 0, 255, 70))
#         newData.append(item)

pred.putdata(newData)
```
